# Pytrends sobre el fin del mundo

Obtener informacion sobre las siguientes tendencias:
- FIn del mundo
- Calentamiento global
- Terremoto
- Tsunami

Importamos la libreria Pytrends

```python
from pytrends.request import TrendReq
```

Inicializamos el objeto TrendReq, creamos la lista con los temas a buscar

```python
pytrends = TrendReq(hl='en-US', tz=360)

kw_list = ["Fin del mundo", "Calentamiento global", "Terremoto", "Tsunami"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
```

Creamos la variable ***'data'*** para almacenar el resultado e importamos la libreria ***plotly*** para poder graficar el resultado

```python
import plotly.express as px

data = pytrends.interest_over_time()
data = data.reset_index()

fig = px.line(data, x= "date" , y=kw_list, title= 'Keyword Web Search Interest Over Time' ) 

fig.show()
```

### Importamos la libreria ***Pandas*** para poder manipular los datos. 

```python
import pandas as pd
```

Creamos un DataFrame para ver las tendencias por Pais

```python
df_by_region = pd.DataFrame(pytrends.interest_by_region(resolution= 'COUNTRY' , inc_low_vol= True , inc_geo_code= False ) )
```

#### Top 10 tendencia ***Fin del Mundo*** por Pais

```python
df_by_region_finMundo = df_by_region[['Fin del mundo']].sort_values(by='Fin del mundo', ascending=False).head(10)
df_by_region_finMundo
```

#### Top 10 tendencia ***Calentamiento Global*** por Pais

```python
df_by_region_calentamiento = df_by_region[['Calentamiento global']].sort_values(by='Calentamiento global', ascending=False).head(10)
df_by_region_calentamiento
```

#### Top 10 tendencia ***Terremoto*** por Pais

```python
df_by_region_terremoto = df_by_region[['Terremoto']].sort_values(by='Terremoto', ascending=False).head(10)
df_by_region_terremoto
```

#### Top 10 tendencia ***Tsunami*** por Pais

```python
df_by_region_tsunami = df_by_region[['Tsunami']].sort_values(by='Tsunami', ascending=False).head(10)
df_by_region_tsunami
```
